<h2> 2.8 Distribution of restaurant's rating according to city

```python
def ci_rating_dist(city_name):
    y=df[df.city==str(city_name)]
    plt.title(f'distribution of restaurant ratings in {city_name}')
    plt.xlabel('Rating')
    return sns.distplot(x=y.aggregate_rating)
    
```

<h2>2.12 Identify and visualize the top restaurant chains based on the number of outlets.
Explore the ratings of these top chains.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate average rating per restaurant
avg_rating = df.groupby('name')['aggregate_rating'].mean().reset_index()

# Optional: sort by rating
avg_rating = avg_rating.sort_values(by='aggregate_rating', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=avg_rating, x='name', y='aggregate_rating', palette='Set3')
plt.xticks(rotation=45)
plt.title('Average Aggregate Rating per Restaurant')
plt.tight_layout()
plt.show()
```
